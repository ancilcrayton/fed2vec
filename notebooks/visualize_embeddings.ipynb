```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from gensim.models.doc2vec import Doc2Vec
from sklearn.manifold import TSNE
%matplotlib inline
```

```python
# Load trained model
model = Doc2Vec.load("../results/fed2vec.model")
```

```python
# Create dataset to reduce dimensionality
X = model[model.wv.vocab]
# See the shape
X.shape
```

```python
# Initialize tSNE and create reduced matrix
tsne = TSNE(n_components=2)
reduced = tsne.fit_transform(X)
```

```python
# Create scatter plot of the words in the two-dimensional vector space
plt.figure(figsize=(20, 20), dpi=80)
plt.scatter(reduced[:, 0], reduced[:, 1])
words = list(model.wv.vocab)
for i, word in enumerate(words):
    plt.annotate(word, xy=(reduced[i, 0], reduced[i, 1]))
plt.show()
```
